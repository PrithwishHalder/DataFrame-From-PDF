```python
import pandas as pd
import pdfplumber
import re
from collections import namedtuple
```

```python
Table1 = namedtuple("Table1","Sr_No Claims_Experience For_Death For_Maturity Survival_Benefit For_Annuities_Pension For_Surrender Other_Benefits Date")
Table2 = namedtuple("Table2","Sr_No Claims_Experience Death Maturity Survival_Benefit Annuities_Pension Surrender Other_Benefits Date")
```

```python
date_re = re.compile(r'(\d{2}\.\d{2}\.\d{4})')
line1_re = re.compile(r'^(\d{1}) ([A-Z][a-zA-Z/ ]+[a-z*]) ([0-9 ]*)')
line2_re = re.compile(r'^([A-Z0-9][a-zA-Z0-9/ ]+[a-z*]) ([0-9 ]*)')
items = line2_re.search('1year and above 0 133 58 296')
item = items.group(2)
print(item.split())
```

```python
list1 = []
list2 = []
list_count = -1

with pdfplumber.open("Q1.pdf") as pdf:
    pages = pdf.pages
    for page in pdf.pages:
        text = page.extract_text()
        for line in text.split('\n'):
            print(line)
            if date_re.search(line):
                date = date_re.search(line).group(0)
                list_count += 1
            elif line1_re.search(line):
                items = line1_re.search(line)
                item = items.group(3).split()
                while len(item) != 6:
                    item.append(0)
                if list_count==0:
                    list1.append(Table1(items.group(1),items.group(2),*item, date))
                elif list_count==1:
                    list2.append(Table2(items.group(1),items.group(2),*item, date))
            elif line2_re.search(line):
                items = line2_re.search(line)
                item = items.group(2).split()
                while len(item) != 6:
                    item.append(0)
                if list_count==0:
                    list1.append(Table1("",items.group(1),*item, date))
                elif list_count==1:
                    list2.append(Table2("",items.group(1),*item, date))
```

```python
df1 = pd.DataFrame(list1)
df2 = pd.DataFrame(list2)
df1
```
